## Histogramas com Gráfico de Distorção

O **Distplot** é usado basicamente para o conjunto univariante de observações e o visualiza por meio de um histograma, ou seja, apenas uma observação e, portanto, escolhemos uma coluna específica do conjunto de dados.


```python
plot = sns.FacetGrid(iris_data, hue="species")
plot.map(sns.distplot, "sepal_length").add_legend()

plot = sns.FacetGrid(iris_data, hue="species")
plot.map(sns.distplot, "sepal_width").add_legend()

plot = sns.FacetGrid(iris_data, hue="species")
plot.map(sns.distplot, "petal_length").add_legend()

plot = sns.FacetGrid(iris_data, hue="species")
plot.map(sns.distplot, "petal_width").add_legend()

plt.show()
```

### Conclusão

* No caso de Comprimento Sepal, há uma grande quantidade de sobreposição.

* No caso da largura do sepal também, há uma grande quantidade de sobreposição.

* No caso de Petal Length, há muito pouca sobreposição.
Também no caso da largura da pétala, há uma pequena quantidade de sobreposição.


### Portanto, podemos usar o comprimento da pétala e a largura da pétala como o recurso de classificação.

## Mapas de calor

O mapa de calor é uma técnica de visualização de dados usada para analisar o conjunto de dados como cores em duas dimensões. Basicamente, ele mostra uma correlação entre todas as variáveis ​​numéricas no conjunto de dados. Em termos mais simples, podemos representar graficamente a correlação encontrada acima usando os mapas de calor.


```python
sns.heatmap(iris_data.corr(method='pearson'), annot = True);

plt.show()
```

### Conclusões

* A largura e o comprimento da pétala têm altas correlações (0,96).

* O comprimento da pétala e o comprimentos da sépala têm boas correlações (0,87).

* A largura da pétala e o comprimento da sépala têm boas correlações (0,82).

### Remover Outliers

* Remover outliers de iris_setosa_data - petal_width

```python
# IQR
Q1 = np.percentile(iris_setosa_data['petal_width'], 25,
                interpolation = 'midpoint')

Q3 = np.percentile(iris_setosa_data['petal_width'], 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1

print("Old Shape: ", iris_setosa_data.shape)

# Upper bound
upper = np.where(iris_setosa_data['petal_width'] >= (Q3+1.5*IQR))

# Lower bound
lower = np.where(iris_setosa_data['petal_width'] <= (Q1-1.5*IQR))

print(upper)
print(lower)

# Removing the Outliers
iris_setosa_data.drop(upper[0], inplace = True)
iris_setosa_data.drop(lower[0], inplace = True)

print("New Shape: ", iris_setosa_data.shape)

sns.boxplot(x='petal_width', data=iris_setosa_data)
```
